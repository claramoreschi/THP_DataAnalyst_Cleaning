## Exercice 1 : Les factures

##### Les bibliotèques utilisées

```python
import pandas as pd
import numpy as np
```

#### Ouverture du csv et conversion pour l'utiliser en dataframe

```python
data = pd.read_csv("/Users/claramoreschi/Desktop/26oct/projet_un/Invoice.csv", encoding='latin-1',  sep="|", engine='python')
df = pd.DataFrame(data)
df
```

### 1) Retirer tout autre caractère (hormis 4 chiffres et 2 lettres) de la colonne SITE_NAME

```python
new_site_name = df['SITE_NAME']
df['SITE_NAME'] = new_site_name.str[-7:-1]
df
```

### 2) Supprimer la colonne INVOICE_NAME (qui ne servira pas)

```python
del df["INVOICE_NAME"]
```

### 3) Ajouter une colonne qui calcule le nombre de jours entre le début de la période de facturation et la fin de la période.

```python
# On transforme en dates.
df["BEGIN"] = df["BEGIN"].apply(pd.to_datetime)
df["END"] = df["END"].apply(pd.to_datetime)
```

```python
# On créée la nouvelle colonne avec le nombre de jours
df['DIFFERENCE_DAYS']= (df['END']- df['BEGIN']).dt.days
```

```python
# Affichage du nouveau tableau
df
```

### 4) À partir de la nouvelle colonne créée DIFFERENCE_DAYS, calculer la moyenne de la consommation journalière dans une nouvelle colonne.

```python
# Je vérifie le type des colonnes pour voir si je peux calculer.
df["KWH_PERIOD"].dtype 
df["DIFFERENCE_DAYS"].dtype

# Ce sont bien des INT, mais je n'arrive pas de suite à les diviser car dans KWH_PERIOD il y a un espace et ça me renvoit une erreur.
```

```python
# Je dois donc supprimer cet espace.
df['KWH_PERIOD'] = pd.to_numeric(df['KWH_PERIOD'].str.replace(' ', ''))
```

```python
# J'affiche la colonne pour vérifier.
df["KWH_PERIOD"] 
# C'est OK !
```

```python
# Je peux maintenant faire la moyenne de consommation journalière.
df['AV_CONSO_PER_DAY'] = df['KWH_PERIOD'] / df['DIFFERENCE_DAYS']
```

```python
# J'affiche le nouveau tableau avec cette nouvelle colonne.
df
```

#### 5) Créer une nouvelle colonne avec le total des factures par antenne.

```python
# Création de la nouvelle colonne où je regroupe les données de mes deux colonnes (GROUBY) et où je fais le total avec TRANSFORM et 'count' 
df['TOTAL_INVOICES'] = df.groupby(['SITE_NAME'])['SITE_NAME'].transform('count')
```

```python
# J'affiche le nouveau tableau
df
```
