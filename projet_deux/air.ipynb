### 1) Supprimer toutes les données qui ne concernent pas la journée du 06/08/21

```python
# Tout d'abord je supprime les caractères qui ne sont pas nécessaires (heures)
new_date = df['Last Updated']
df['Last Updated'] = new_date.str[:-15]

# J'affiche le nouveau tableau
df
```

```python
# Je convertis les données de 'Last Uptaded' qui sont actuellement des objets (string) en date.
df["Last Updated"] = df['Last Updated'].astype('datetime64[ns]')
df["Last Update"] = df["Last Updated"].apply(pd.to_datetime, format = '%Y/%m/%d') 
```

```python
# Je vérifie qu'à présent le type soit bien une date et non plus une string 
df["Last Updated"].dtype
```
